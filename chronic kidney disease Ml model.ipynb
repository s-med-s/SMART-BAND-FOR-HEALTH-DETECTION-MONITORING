```python
import numpy as np
import pandas as pd
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r'C:\\Users\\HP\\kidney_disease.csv')
df
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
label_encoder = LabelEncoder()
categorical_cols = ['rbc', 'pc', 'pcc', 'ba', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col].astype(str))

print("Data types after encoding:", df.dtypes)
```

```python
numeric_cols = df.select_dtypes(include=['number'])
df[numeric_cols.columns] = numeric_cols.apply(lambda x: x.fillna(x.mean()))

# طباعة عدد القيم المفقودة بعد المعالجة
print("Missing values after filling:", df.isnull().sum())
```

```python
selected_features = ['hemo',  'sg',  'bp', 'age', 'htn', 'bgr','bu' ,'sc','classification']
df_selected = df[selected_features]

X = df_selected.drop(columns=['classification'])
y = df_selected['classification']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)

# تطبيق التطبيع على بيانات التدريب


# تدريب نموذج Random Forest على البيانات المطَبعة
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)

# التنبؤ باستخدام بيانات الاختبار
y_pred = rf_classifier.predict(X_test)

# يمكنك متابعة حساب دقة النموذج أو مصفوفة الالتباس
```
